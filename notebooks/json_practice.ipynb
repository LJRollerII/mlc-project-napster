```python
import json
import pandas as pd
```

```python
# Opening JSON file
g = open('../data/genre.json')
 
# returns JSON object as a dictionary
genre_json = json.load(g)

# since the value lists have varying lengths, we have to use a lambda function to iterate over the keys and values as a series
genres = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in genre_json.items()]))

genres.head(20)
```

```python
#Data is wide and we would like it to be long. We need to reshape the DataFrame (flip columns and rows)
```

```python
genres = pd.melt(genres)
```

```python
genres.head(20)
```

```python
#Remove null values
genres = genres.dropna()
```

```python
genres.head(20)
```

```python
#Clean up Pop - normalize
#Reset index
```

```python
genres = genres.reset_index(drop=True)
genres.head(20)
```

```python
#rename columns
genres.columns = ['artist', 'genre']
genres.head(20)
```
