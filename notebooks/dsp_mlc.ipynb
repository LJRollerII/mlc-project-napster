```python
no_writer_dsp = no_writer.groupby(['dsp'])['streams'].sum().sort_values(ascending=False).nlargest(5).to_frame()
no_writer_dsp.reset_index(inplace=True)
ax = sns.barplot(x=no_writer_dsp.dsp, y=no_writer_dsp.streams, data=no_writer_dsp, palette="Greys_r", edgecolor='none')
ax.ticklabel_format(style='plain', axis="y")
plt.xticks(rotation=65)
plt.ylabel('STREAMS')
plt.xlabel('DSP')
plt.title('DSPs Missing Songwriter Info',
          fontsize=12,
          backgroundcolor="black",
          pad='12.0',
          color="white",
          fontstyle='oblique')
ax.bar_label(ax.containers[0], fmt = '%d', 
             #padding=2, 
             label_type='center', 
             color='white', 
             backgroundcolor='black')
#ax.axis("off")
right_side = ax.spines["right"]
right_side.set_visible(False)
left_side = ax.spines["left"]
left_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
bottom_side = ax.spines["bottom"]
bottom_side.set_visible(False)
plt.savefig('DSP Labels.png', transparent=False, bbox_inches="tight")
```
